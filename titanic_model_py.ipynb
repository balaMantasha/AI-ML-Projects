```python
#no of males and no of males survivrd and died
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('train.csv')
males = df[df['Sex'] == 'male']
print(males.shape[0])

males_survived = df[(df['Sex'] == 'male') & (df['Survived'] ==1)]
print(males_survived.shape[0])

males_died = df[(df['Sex'] == 'male') & (df['Survived'] ==0)]
print(males_died.shape[0])

status = ['Survivrd' , 'Died']
sizes =[males_survived.shape[0],males_died.shape[0]]
plt.pie(sizes, labels = status ,  autopct = '%1.1f%%')
```

```python
import pandas as pd
df = pd.read_csv('train.csv')
df
```

```python
# no of passengers in different classes

class1 = df[df['Pclass'] == 1]
print('Number of passengers in Class 1 was',class1.shape[0])

class2 = df[df['Pclass'] == 2]
print('Number of passengers in Class 2 was',class2.shape[0])

class3 = df[df['Pclass'] == 3]
print('Number of passengers in Class 3 was',class3.shape[0])
```

```python
#no of survivors in ist class 2nd class 3rd class

istclass_survivors = df[(df['Pclass'] ==1)  & (df['Survived'] == 1)]
print(istclass_survivors.shape[0])

Secondclass_survivors = df[(df['Pclass'] ==2)  & (df['Survived'] == 1)]
print(Secondclass_survivors.shape[0])

Thirdclass_survivors = df[(df['Pclass'] ==3)  & (df['Survived'] == 1)]
print(Thirdclass_survivors.shape[0])




# stacked bar grapph
categories = ['Ist class', '2nd class', '3rd class']
values1 = [class1.shape[0], class2.shape[0], class3.shape[0]]
values2 = [istclass_survivors.shape[0],Secondclass_survivors.shape[0], Thirdclass_survivors.shape[0], ]

# Plot the stacked bar graph
plt.bar(categories, values1, color='skyblue', label='Total passengers')
plt.bar(categories, values2, bottom=values1, color='lightgreen', label='Survived')

# Add labels and legend
plt.title('Survival across passenger classes')
plt.xlabel('Passenger classes')
plt.ylabel('number of passengers')
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('train.csv')

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# average age with missing values
average_age_1 = df['Age'].mean()
print("Average age of people onboard was ", average_age_1)
```

```python
# number of missing vlaues in age
missing_values_in_age = df['Age'].isnull().sum()
print(missing_values_in_age)
```

```python
# filling missing values in age
df['Age'].fillna(average_age_1, inplace=True)

```

```python
# Print the correlation matrix
numeric_data = df.select_dtypes(include=['float64' , 'int64'])
correlation_matrix = numeric_data.corr()

# Print the correlation matrix
print("Correlation Matrix for Float Columns:")
print(correlation_matrix)
```

```python
# heat map
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Passengers travelling with family
pass_with_family= df[(df['Parch'] > 0)  | (df['SibSp'] > 0)]

print(pass_with_family.shape[0])

# without family
pass_withOut_family= df[(df['Parch'] == 0)  & (df['SibSp'] == 0)]
print(pass_withOut_family.shape[0])

survived_w_family = pass_with_family[pass_with_family['Survived'] == 1].shape[0]
percentage_1=(survived_w_family/pass_with_family.shape[0])*100
print(percentage_1)

survived_wo_family = pass_withOut_family[pass_withOut_family['Survived'] == 1].shape[0]
print(survived_wo_family)
percentage_2=(survived_wo_family/pass_with_family.shape[0])*100
print(percentage_2)

# bar graphs
xaxis=['passengers with family' , 'Passengers travelling alone']
values=[percentage_1, percentage_2]

plt.bar(xaxis, values, color='skyblue')
plt.title('Survival of passengers travelling with and without family')

plt.ylabel('Values(percentage)')
plt.show()
```
